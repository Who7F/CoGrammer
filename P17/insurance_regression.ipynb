```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
source = pd.read_csv('insurance.csv')
source.head()
```

I would like to know what source['age'].values is not the same as source.iloc[:,:1].values

```python
x = source.iloc[:,:1].values
y = source.iloc[:,-1].values

plt.scatter(x,y)
plt.ylabel("Charges")
plt.xlabel("Age")
plt.show()
```

```python
sexy_data_model = LinearRegression()
sexy_data_model.fit(x,y)
charges_pred = sexy_data_model.predict(x)
```

```python
plt.scatter(x, y, color = 'b')
plt.plot(x,charges_pred, color = 'y')      
plt.ylabel("Charges")
plt.xlabel("Age")
plt.show()
```
