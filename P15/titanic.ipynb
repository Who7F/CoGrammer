EDA on on the Titanic dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic.csv")
df.head()
```

Get the list of data types

```python
df.dtypes
```

Identified fields of interest as: Pclass, Sex, Age, sibsp, parch, Embarked

sibsp - Number of Siblings/Spouses Aboard. 
parch - Number of Parents/Children Aboard.

By Class

```python
first_class, second_class, thired_cass = df.groupby('Pclass')['Survived'].mean()
plt.bar(["First Class", "Second Class", "Thired Cass"],[first_class, second_class, thired_cass])
plt.ylabel("Survived %")
plt.xlabel("Classes")
plt.show()
```

```python
By gender
```

```python
female, male = df.groupby('Sex')['Survived'].mean()
plt.bar(['Female', 'Male'],[female, male])
plt.ylabel("Survived %")
plt.xlabel("Sex")
plt.show()
```

```python
By age groups
```

```python
age_bins = [0, 10, 17, 34, 55, 150]
age_labels = ['0-10','11-17','18-34','35-54','55+']
df['age_groups'] = pd.cut(df['Age'], bins = age_bins, labels = age_labels)
    
plt.bar(age_labels, df.groupby('age_groups')['Survived'].mean())
plt.ylabel("Survived %")
plt.xlabel("Age Groups")
plt.show()
```

```python
a,b,c,d,e,f,g = df.groupby('SibSp')['Survived'].mean()
plt.bar(["0", "1", "2", "3", "4", "5", "8"],[a,b,c,d,e,f,g])
plt.ylabel("Survived %")
plt.xlabel("Siblings/Spouses")
plt.show()
```

```python
a,b,c,d,e,f,g = df.groupby('Parch')['Survived'].mean()
plt.bar(["0", "1", "2", "3", "4", "5", "6"],[a,b,c,d,e,f,g])
plt.ylabel("Survived %")
plt.xlabel("Parents/Children")
plt.show()
```

```python
c_emb, q_emb, s_emb = df.groupby('Embarked')['Survived'].mean()
plt.bar(["C", "Q", "S"],[c_emb, q_emb, s_emb])
plt.ylabel("Survived %")
plt.xlabel("Embarked")
plt.show()
```
